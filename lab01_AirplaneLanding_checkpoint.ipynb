## Discussion of the problem

### Calculating the acceleration
The acceleration is defined as the ratio between the change $\Delta v$ in velocity and duration $\Delta t$ of the time lapse during which such change in velocity occurs.
    
$$a = \dfrac{\Delta v}{\Delta t} = \dfrac{v_f - v_i}{t_f - t_i}$$
    
To describe the motion of the airplane along the runway we choose the an axis, parallel to the runway, oriented in the same relative direction in which the airplane arrives.

It should be pointed out that the final speed is zero and the initial speed is positive, so
* the speed variation is negative and
* the acceleration will be negative

$$ a = \dfrac{-v_0}{\Delta t} $$

&nbsp;


### Calculating the stopping distance
We can evaluate the stopping distance using the equation for the motion with constant acceleration

$$ x = x_0 + v_0\cdot(t - t_0) + \dfrac{1}{2}a\cdot(t - t_0)^2 $$

In the present case, the values of  $x_0, t_0$ and $v_0$ are known. Moreover, the value of the acceleration $a$ was calculated in the first part of the problem. To calculate the value of the coordinate $x$ for which the airplane achieves a complete stop it will sufficient to substitue for $(t-t_0)$ the value given in the text of the problem.

The distance occurring between the touch down position and the final stop of the airplane is what we mean as  **stopping distance**. We have thus

$$ d = x - x_0 $$

i.e.

$$ d = v_0\cdot(t - t_0) + \dfrac{1}{2}a\cdot(t - t_0)^2 $$

&nbsp;

Instead of substituting for the numerical value of $a$ we can carry out a symbolic calculation. Considering the expression for the acceleration $a$, we have

$$ d = v_0 \Delta t + \dfrac{1}{2}\dfrac{\Delta v}{\Delta t}(\Delta t)^2 $$

Since in this case $\Delta v = 0 - v_0$, it follows

$$ d = v_0 \Delta t - \dfrac{1}{2}v_0 \Delta t $$

then

$$ d = \dfrac{1}{2}v_0 \Delta t $$

### An alternative method
Considering the relation
$$ v^2 = v_0^2 + 2 a\cdot(x - x_0) $$

in this case we obtain

$$ x - x_0 = \dfrac{-v_0^2}{2 a} $$

The value of the acceleration $a$ is known from the first part of the problem, and $v_0$ is given as a datum. Thus, with this equation we can calculated the stopping distance $d$ which is equal to $(x-x_0)$.

Looking at the equation above, we can remark that

- the stopping distance is directly proportional to the square of the initial velocity

- the stopping distance is inversely proportional to the magnitude of the acceleration, which depends, in turn, on the intensity of the breaking action

- The variable $d = x - x_0$ has positive values, since the factor  $-v_0^2$ is negative and the acceleration $a$ is also negative

&nbsp;

In the other hand, symbolic calculus gives

$$ x - x_0 = -\dfrac{v_0^2}{2} \dfrac{\Delta t}{\Delta v} $$

thus,

$$ x - x_0 = -\dfrac{v_0^2}{2} \dfrac{\Delta t}{-v_0} $$

and, finally,

$$ x - x_0 = \dfrac{1}{2} v_0 \Delta t $$
